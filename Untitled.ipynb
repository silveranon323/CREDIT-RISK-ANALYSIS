```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset 
#datapreprocessing - checking for missing values,feature analysis
df = pd.read_csv("german_credit_data.csv")
df.head()
df.info()
```

```python
df.isnull().sum() #checking for the missing values. Part of data preprocessing
df.isnull().mean()
```

```python
categorical_features = df.select_dtypes(include=[object]).columns
print(categorical_features) #printing columns that have categorical data sets
# we need to convert categorical value into numeric value.
```

```python
def SavingChecking_LabelEncoder(value):
    if value == "little":
        return 1
    elif value == "moderate":
        return 2
    elif value == "quite rich":
        return 3
    elif value == "rich":
        return 4
    else:
        return 0
```

```python
cols = ['Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose', 'Risk']

for col in cols:
    line = col + " : "
    for v in df[col].unique():
        line = line + str(v) + " , "
    display(line)
```

```python
df["Saving accounts"] = df["Saving accounts"].apply(SavingChecking_LabelEncoder)
df["Checking account"] = df["Checking account"].apply(SavingChecking_LabelEncoder)
```

```python
df
```

```python
# Convert to Numerical Value

def House_LabelEncoder(value):
    if value == "free":
        return 0
    elif value == "rent":
        return 1
    elif value == "own":
        return 2
```

```python

df["Housing"] = df["Housing"].apply(House_LabelEncoder)
```

```python
df
```

```python
# Data Pre-processing For Discrete Categorical Columns (Sex,Purpose,Risk)

from sklearn.preprocessing import LabelEncoder
for col in ["Sex", "Purpose", "Risk"]:
    le = LabelEncoder()
    le.fit(df[col])
    df[col] = le.transform(df[col])
```
